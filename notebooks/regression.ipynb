```python
import sys
sys.path.append("../")
from ml import regression
import matplotlib.pyplot as plt
```

```python
d = regression.ArtificialDataset()
d.generate_X_gaussian(2, 1000)
weights, snr = d.generate_y_linear()

r = regression.Manager(2)
r.load_data(d.X, d.y)

print(weights)
print(r.feature_names)
print()
print(regression.describe_data(r))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.tight_layout()

regression.plot_data_one_feature(ax1, r, 0)
regression.plot_data_one_feature(ax2, r, 1)
plt.show()
```

```python
fig, ax = plt.subplots()
regression.plot_data_two_features(
    ax, r, 0, 1)
plt.show()
```

```python
r.clear_models()
r.add_model(regression.ModelLinear(var_e = 100, var_ratio=5))
r.add_model(regression.ModelLinear(var_ratio=0.1))
r.add_model(regression.ModelLinear(var_ratio=1))

r.fit()
print(regression.describe_results(r))

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.tight_layout()
regression.plot_results_one_feature(ax1, r, 0)
regression.plot_results_one_feature(ax2, r, 1)
plt.show()
```

```python
fig1, ax1 = plt.subplots()
fig2, (ax2, ax3) = plt.subplots(1, 2, figsize=(9, 4))
fig2.tight_layout()

regression.plot_results_two_features(ax1, ax2, ax3, r, 0, 1)
plt.show()
```
