```python
import sys
sys.path.append("../")
from ml import classification
import matplotlib.pyplot as plt
```

```python
D=2
C=3

d = classification.ArtificialDataset(C)
d.generate_X_gaussian(D, 1000)
W = d.generate_y_linear()

cm = classification.Manager(D, C)
cm.load_data(d.X, d.y)

print(W)
print(cm.feature_names)
print()
print(classification.describe_data(cm))
```

```python
classification.plot_data_two_features(None, cm, 0, 1)
plt.show()
```

```python
cm.clear_models()
cm.add_model(classification.ModelSimple(
    learning_rate=1, var_w=0.1))
cm.add_model(classification.ModelSimple(
    learning_rate=1, var_w=1))
cm.add_model(classification.ModelSimple(
    learning_rate=1, var_w=10))
cm.add_model(classification.ModelSimple(
    transform="gaussian_radial_basis"))

cm.fit()
print(classification.describe_results(cm))
```

```python
classification.plot_results_two_features(None, cm, 0, 1)
```
